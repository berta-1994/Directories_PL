## Geolocation Comparison:

I compare the total geolocated points in both models used. Tesseract and AI OCR using gemini.

### Tesseract

```python
# Iteració:
ratios={}
for element in city_list:
  file_path = f'geo_files/{element}/geo_{element}.csv'
  if os.path.exists(file_path):
    geo_city = pd.read_csv(f'geo_files/{element}/geo_{element}.csv')

    # Sum all full addresses which are not na
    geo_addresscount = geo_city['Full_Address'].notna().sum()
    geo_located = geo_city['Latitude'].count()
    print(f"City: {element}, Number of addresses: {geo_addresscount}, Number of georeferenced points: {geo_located}, ratio: {geo_located/geo_addresscount}")
    ratios.update({element:geo_located/geo_addresscount})
  else:
    print(f"City: {element} not found")
    continue
```

### AI Model

```python
# Iteració  AI OCR
ratios={}
for element in city_list:
  file_path = f'ai_files/{element}/geo_{element}.csv'
  if os.path.exists(file_path):
    geo_city = pd.read_csv(f'ai_files/{element}/geo_{element}.csv')

    geo_addresscount = geo_city['full_address'].notna().sum()
    geo_located = geo_city['Lat'].count()
    print(f"City: {element}, Number of addresses: {geo_addresscount}, Number of georeferenced points: {geo_located}, ratio: {geo_located/geo_addresscount}")
    ratios.update({element:geo_located/geo_addresscount})
  else:
      print(f"City: {element} not found")
      continue
```

### Unique building Names

```python
# TESSERACT
# For each city find unique building names
empty_buildings = []

for city in city_list:
  count = 0
  file_path = f'geo_files/{city}/geo_{city}.csv'
  if os.path.exists(file_path):
    geo_city = pd.read_csv(f'geo_files/{city}/geo_{city}.csv')
    if geo_city['Building'].notnull().any() and geo_city['Latitude'].isna().any():
      unique_buildings = geo_city['Building'].unique()
      count = len(unique_buildings)
      year = 1912
      # print(f"City: {city}, Unique building names: {unique_buildings}")
      empty_buildings.append([city, year, unique_buildings, count])
  else:
    print(f"City: {city} not found")
    continue
empty_buildings = pd.DataFrame(empty_buildings, columns=["city","year", "unique_buildings", "count"])
empty_buildings
```

### AI

```python
# AI MODEL
# For each city find unique building names

empty_buildings = []

for city in city_list:
  count = 0
  file_path = f'ai_files/{city}/geo_{city}.csv'
  if os.path.exists(file_path):
    geo_city = pd.read_csv(f'ai_files/{city}/geo_{city}.csv')
    if geo_city['building'].notnull().any() and geo_city['Lat'].isna().any():
      unique_buildings = geo_city['building'].unique()
      count = len(unique_buildings)
      year = 1912
      # print(f"City: {city}, Unique building names: {unique_buildings}")
      empty_buildings.append([city, year, unique_buildings, count])
  else:
    print(f"City: {city} not found")
    continue
empty_buildings = pd.DataFrame(empty_buildings, columns=["city", "year", "unique_buildings", "count"])
empty_buildings
```

```python
# Creaning dataframe of unlocated buildings:
# each element in the unique builiding list of each city should be a single row.

empty_buildings = empty_buildings.explode('unique_buildings').reset_index(drop=True)
# removing space at beginning and leading periods

empty_buildings['unique_buildings'] = empty_buildings['unique_buildings'].str.strip().str.replace(r'^\.', '', regex=True)

# Sorting by 'city' and 'cleaned_buildings'
empty_buildings = empty_buildings.sort_values(by=['city', 'unique_buildings'])

empty_buildings
empty_buildings.to_csv('other/AI_empty_bnames.csv', index=False)
```

```python
# Total geolocated elements
data = []
for city in city_list:
  file_path = f'ai_files/{city}/geo_{city}.csv'
  if os.path.exists(file_path):

    geo_city = pd.read_csv(f'ai_files/{city}/geo_{city}.csv')
    address = geo_city['address'].count()
    building = geo_city['building'].count()
    full_address = geo_city['full_address'].count()
    geoloc = geo_city['Lat'].count()
    total = len(geo_city)
    ratio = round((geoloc/total)*100,2)
    data.append([city, address, building, full_address, geoloc, ratio])
  else:
    print(f"City: {city} not found")
    continue

city_map = pd.DataFrame(data, columns=["city", "address", "building", "full_address", "geoloc", "ratio"])
city_map
city_map.to_csv('other/AI_city_summary.csv', index=False)
```
