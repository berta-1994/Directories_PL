```python
cities = pd.read_excel('City_list_coreUS.xlsx')
```

```python
cities.head()
```

```python
# create a list with all cities:
city_list = []
for element in cities['metro']:
  city_list.append(element)

print(city_list)
```

```python

cwd = os.getcwd()
cwd
```

```python
# Create folders with all cities in the entry , cropped and temp folders
for element in city_list:
  if path.exists(cwd + '/entry/' + element) == False:
    os.mkdir(cwd + '/entry/' + element)

for element in city_list:
  if path.exists(cwd + '/temp/' + element) == False:
    os.mkdir(cwd + '/temp/' + element)

for element in city_list:
  if path.exists(cwd + '/cropped/' + element) == False:
    os.mkdir(cwd + '/cropped/' + element)

for element in city_list:
  if path.exists(cwd + '/geo_files/' + element) == False:
    os.mkdir(cwd + '/geo_files/' + element)

for element in city_list:
  if path.exists(cwd + '/files/' + element) == False:
    os.mkdir(cwd + '/files/' + element)
```
